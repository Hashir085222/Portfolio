```python
import tensorflow as tf
from tensorflow.keras import layers, models

# 1. Load the training dataset
# label_mode='binary' forces target.shape to be (None, 1)
train_data = tf.keras.utils.image_dataset_from_directory(
    'face_detection_dataset/train_data/train_data',
    image_size=(128, 128),
    batch_size=32,
    label_mode='binary' 
)

# 2. Load the test dataset (make sure it also has it_sme and it_snotme inside)
test_data = tf.keras.utils.image_dataset_from_directory(
    'face_detection_dataset/train_data/train_data',
    image_size=(128, 128),
    batch_size=32,
    label_mode='binary'
)

# 3. Build the CNN Model
model = models.Sequential([
    # Standardize image pixel values from [0, 255] to [0, 1]
    layers.Rescaling(1./255, input_shape=(128, 128, 3)),
    
    # CNN Layers
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    # Dense Layers
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    
    # Output layer: 1 output unit with 'sigmoid' to match target.shape (None, 1)
    layers.Dense(1, activation='sigmoid')
])

# 4. Compile with binary_crossentropy loss
model.compile(
    optimizer='adam',
    loss='binary_crossentropy', # Crucial: matches output.shape (None, 1)
    metrics=['accuracy']
)

# 5. Train the model
history = model.fit(
    train_data,
    validation_data=test_data,
    epochs=15  
)
```

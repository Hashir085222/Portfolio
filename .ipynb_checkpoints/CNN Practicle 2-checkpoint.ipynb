```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
import numpy as np
import cv2
```

```python
train_gen = ImageDataGenerator(rescale=1./255, validation_split= 0.25)
```

```python
train_data = train_gen.flow_from_directory(
    "facerecognitiondataset/train_data/train_data/",
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset= 'training'
    
)
validation_data = train_gen.flow_from_directory(
    "facerecognitiondataset/train_data/train_data/",
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset= 'validation'
    
)



# test_data = test_gen.flow_from_directory(
#     "facerecognitiondataset/test_data/test_data/",
#     target_size=(128, 128),
#     batch_size=32,
#     class_mode='categorical'
# )
```

```python
model = Sequential()
model.add(Conv2D(124,(3,3),activation= 'relu', input_shape = (128,128,3)))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(64,(3,3), activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(64,(3,3), activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(32,(3,3), activation='relu'))
model.add(MaxPooling2D((2,2)))
```

```python
model.summary()
```

```python
model.add(Flatten())
model.add(Dense(64,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(2,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
             loss = 'categorical_crossentropy',
             metrics= ['accuracy'],
             )
```

```python
model.fit(train_data,epochs=10,verbose=1,validation_data=validation_data)
```

```python
model.save('models/facerecognition.keras')
```

```python
model = load_model('models/facerecognition.keras')
```

```python
label_map = train_data.class_indices
inv_label_map = {v: k for k, v in label_map.items()}
print("Label mapping:", inv_label_map)
```

```python
def predict_image(path):
    img = cv2.imread(path)
    img = cv2.resize(img, (128, 128))
    img = img / 255.0
    img = np.expand_dims(img, axis=0)

    preds = model.predict(img)
    index = np.argmax(preds)          # no if/else
    label = inv_label_map[index]      # label from mapping

    print("\nPrediction Scores:", preds)
    print("Predicted Label:", label)
```

```python
predict_image('facerecognitiondataset/train_data/train_data/it_sme/1 (1).jpg')
```
